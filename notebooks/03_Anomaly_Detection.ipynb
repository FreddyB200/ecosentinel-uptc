# 03 - Detección de Anomalías

Análisis y detección de anomalías en patrones de consumo energético.

**Métodos:**
- Z-Score para anomalías estadísticas
- Pattern mining para ineficiencias operativas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys

sys.path.insert(0, '../backend')
from api.ml.anomaly_detector import detect_zscore_anomalies, detect_pattern_anomalies
```

```python
df = pd.read_csv('../backend/data/consumos_uptc.csv', parse_dates=['timestamp'])

# Filtrar una sede/sector para análisis
sede = 'Tunja'
df_tunja = df[df['sede'] == sede].copy()
df_tunja['energia_kwh'] = df_tunja['energia_comedor_kwh']
print(f'Registros: {len(df_tunja)}')
```

```python
# Detectar anomalías Z-Score
anomalias = detect_zscore_anomalies(df_tunja, column='energia_kwh', threshold=2.5)
print(f'Anomalías detectadas: {len(anomalias)}')
anomalias.head()
```
