# 01 - Preparación de Datos

Exploración y limpieza del dataset de consumo energético UPTC.

**Dataset:** 270,000 registros horarios (2018-2025)
**Sedes:** Tunja, Duitama, Sogamoso, Chiquinquirá
**Sectores:** Comedores, Salones, Laboratorios, Auditorios, Oficinas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-darkgrid')
pd.set_option('display.max_columns', 20)
```

```python
# Cargar dataset
df = pd.read_csv('../backend/data/consumos_uptc.csv', parse_dates=['timestamp'])
print(f'Registros: {len(df):,}')
print(f'Período: {df.timestamp.min()} → {df.timestamp.max()}')
df.head()
```

```python
# Info general
df.info()
df.describe()
```

```python
# Distribución por sede
df['sede'].value_counts().plot(kind='bar', title='Registros por Sede')
plt.tight_layout()
plt.show()
```
